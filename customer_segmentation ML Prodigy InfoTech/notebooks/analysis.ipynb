# Mall Customer Segmentation Project

This project applies **K-Means Clustering** to segment mall customers based on their Annual Income and Spending Score.

## Goal
Group customers into distinct clusters to help businesses understand customer behavior and target marketing strategies.

## Dataset
- `Mall_Customers.csv` — contains customer data including Age, Annual Income (k$), and Spending Score (1-100).

## Pipeline
1. Load and preprocess data
2. Select features: `Annual Income` and `Spending Score`
3. Train K-Means model with k=5
4. Visualize clusters

```python
import sys
sys.path.append('..')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

print('Libraries loaded successfully')
```

```python
# Load data
df = pd.read_csv('../data/Mall_Customers.csv')
print('Shape:', df.shape)
df.head()
```

```python
# Basic info
df.info()
df.describe()
```

```python
# Select features
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]
X.head()
```

```python
# Elbow method to find optimal k
wcss = []
for i in range(1, 11):
    km = KMeans(n_clusters=i, random_state=42)
    km.fit(X)
    wcss.append(km.inertia_)

plt.figure(figsize=(8, 4))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.tight_layout()
plt.show()
```

```python
# Train K-Means with k=5
model = KMeans(n_clusters=5, random_state=42)
y_kmeans = model.fit_predict(X)
print('Cluster labels:', np.unique(y_kmeans))
```

```python
# Visualize clusters
colors = ['red', 'blue', 'green', 'cyan', 'magenta']
labels = ['Cluster 1', 'Cluster 2', 'Cluster 3', 'Cluster 4', 'Cluster 5']

plt.figure(figsize=(9, 6))
for i in range(5):
    plt.scatter(
        X.values[y_kmeans == i, 0],
        X.values[y_kmeans == i, 1],
        s=80, c=colors[i], label=labels[i]
    )

plt.scatter(
    model.cluster_centers_[:, 0],
    model.cluster_centers_[:, 1],
    s=200, c='yellow', marker='*', label='Centroids'
)

plt.title('Customer Segments (K-Means, k=5)')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.tight_layout()
plt.savefig('../static/cluster_plot.png')
plt.show()
print('Plot saved to static/cluster_plot.png')
```
